```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python

fh = open('season_1/training_data/poi_data/poi_data')
import re
SEPS = ('\t', ':')
rsplit = re.compile("|".join(SEPS)).split
content = []
for  line in  fh.readlines(): 
    # print (line)
    a = [s.strip() for s in rsplit(line)]
    b = dict(zip(a[1::2], a[2::2]))
    b['name'] = a[0]
    content.append(b)
poi_data = pd.DataFrame(content)
```

```python
cluster_map = pd.read_table('season_1/training_data/cluster_map/cluster_map', delimiter='\t',header = None)
cluster_map.columns = ["name", "mapID"]
```

```python
region_data_1 = pd.concat([cluster_map, poi_data], axis = 1);region_data_1.head()
```
